```python
import requests
from bs4 import BeautifulSoup
import urllib.request
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
```

```python
source = requests.get('https://www.gettyimages.in/photos/royalty-free-images-virat-kohli', headers = headers).text
soup = BeautifulSoup(source, 'lxml')
```

```python
Images=[]
img_links=soup.select('img[src^="https://media.gettyimages.com/id"]')
for i in range(len(img_links)):
    Images.append(img_links[i]['src'])
```

```python
for i in range(len(Images)):
    name="C:/Users/Asus/Downloads/"+str(i)+".jpg"
    urllib.request.urlretrieve(Images[i], name)
```
